```python
# Definindo sets, sets nao podem ser repetidos
classmate = {"alex", "andre","lombarde","andre"}
friends = {"costa", "Lombarde"}
```

```python
# imprimindo uniao
uniao = classmate.union(friends)
print(uniao)
# imprimindo intersecao
print(friends.intersection(uniao))
# imprimindo diferença (mostra os itens presentes na primeira lista"classmate" e nao presentes na segunda"friends")
print(classmate.difference(friends))
```
