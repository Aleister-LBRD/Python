```python
# IF - Estrutura basica do IF
valor = 7

if valor <= 4:
    print ("hello world!")
elif valor == 5:
    print ("igual a 5")
else:
    print ("not good!")
```

```python
# IF - Estrutura de IF aninhado
nome = 'alex'
snome = 'silva'
idade = 36
if nome == 'alex' and snome == 'silva':
    if idade == 35:
        print('ok')
    else:
        print('ainda não')
else: 
    print('no way')
```

```python
# LOOP FOR - Estrutura LOOP FOR com range 
for i in range(0,10):
    print(i)
```

```python
# LOOP FOR - Estrutura LOOP FOR apresentando numeros pares, detalhe no modulo que é o resta da divisao
lt = [2,3,4,5,6,7,8]
for i in lt:
    if i %2 == 0:
        print(i)
```

```python
# LOOP FOR - Criando lista com o range e LOOP FOR
for i in range (0,10):
    lt.append(i)
print(lt)
lt.clear()
```

```python
# LOOP FOR - Criando lista e imprimindo numeros pares com modulo %
for i in range (0,10):
    lt.append(i)
for i in lt:
    if i %2 == 0:
        print(i)
lt.clear()
```

```python
# LOOP FOR - Criando lista e somando a variavel
for i in range (0,6):
    lt.append(i)
print(lt)
soma = 0
for i in lt:
    soma += i
print(soma)
lt.clear()
```

```python
# LOOP FOR - pulando uma etapa e continuando execução
for i in "python":
    if i == 'h':
        continue
    print(i)
```

```python
# LOOP WHILE - Estrutura
i = 0
while i < 10:
    print(i)
    i = i + 1
else:
    print('loop concluido')
```

```python
# LOOP WHILE - com if e break
i = 0
while i < 100:
    if i == 5:
        break
    else:
        pass
    print(i)
    i = i + 1
```

```python
# Metodos - Definindo funcao
def split(text):
    return text.split(" ")
```
