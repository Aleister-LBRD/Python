```python
# Importanto Biblioteca
import PyPDF2 as pdf
```

```python
# Abrindo Arquivo em modo binário
pdf_file = open('AL.pdf', 'rb')

dados = pdf.PdfFileReader(pdf_file)
```

```python
# selecionando a pagina e extraindo texto e atribuindo em variavel
pag1 = dados.getPage(0)
txt_pag1 = pag1.extractText()
print(txt_pag1)
```

```python
# juntando dois arquivos pdf
arq1 = open('AL.pdf', 'rb')
arq2 = open('SI.pdf', 'rb')

dados1 = pdf.PdfFileReader(arq1)
dados2 = pdf.PdfFileReader(arq2)

merge = pdf.PdfFileMerger()

merge.append(dados1)
merge.append(dados2)
merge.write('file_merged.pdf')
```
