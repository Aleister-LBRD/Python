```python
#Definindo a variavel
nome = "Alexandre Lombarde"
```

```python
#impressão com quebra
print ("Alexandre\nLombarde")
```

```python
#soma de caracteres com slice
nome [0]+nome [1]+nome [2]
```

```python
#slice de tudo até 3º caracter
nome [:3]
```

```python
#slice de tudo menos os 3 ultimos
nome [:-3]
```

```python
#slice pulando caracter
nome [::2]
```

```python
#slice invertendo texto
nome [::-1]
```

```python
#Todos os caracteres maiusculos
nome.upper()
```

```python
#Todos os caracteres minusculos
nome.lower()
```

```python
#cortar texto
nome.split("e")
```

```python
#Contar Caracteres
nome.count("a")
```

```python
#Encontrar Caracteres
nome.find("e")
```
