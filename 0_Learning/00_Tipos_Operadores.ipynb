Operadores Aritimeticos
+ [+] soma
+ [-] subtração
+ [*] multiplicacao
+ [/] divisao
+ [%] modulo [sobra da divisão]
+ [**] potencia
+ [//] divisão inteira
+ [int()] converte para inteiro
+ [float()] converte para float

Operadores Relacionais
+ [==] igualdade
+ [!=] desigualdade
+ [>] maior
+ [<] menor
+ [>=] maior ou igual
+ [<=] menor ou igual


Operadores de Atribuicao
+ [=] atribuicao
+ [+=] soma
+ [-=] subtracao
+ [*=] multiplicacao
+ [/=] divisao
+ [%=] modulo
+ [**] potencia
+ [//] divisao inteira


Operadores Logicos
+ [and] ambos operadores True
+ [or] um dos operadores True
+ [not] inverso da logica True

Funcao BUILT IN
+ [str] converter em string
+ [int] converter em integer
+ [float] converter em float
+ [len] tamanho
+ [max] maximo
+ [min] minimo
+ [sum] soma

Manipulando arquivos
+ [open()] abrir arquivo
+ [read()] ler arquivo
+ [write()] escrever arquivo
+ [seek()] retorna o inicio do arquivo
+ [readlines()] lista de linhas do arquivo
+ [close()] fecha o arquivo
+ [map()] adiciona uma sequencia a uma função

Bibliotecas para importação
+ pandas - Planilhas
+ os - Manipulação do SO
+ urllib.request - Buscar URL para scrapping
+ random - Randomização
