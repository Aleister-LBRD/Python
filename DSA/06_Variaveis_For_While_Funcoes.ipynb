<font color='red'>- IF

```python
# IF - Estrutura basica do IF
valor = 7

if valor <= 4:
    print ("hello world!")
elif valor == 5:
    print ("igual a 5")
else:
    print ("not good!")
```

```python
# IF - Estrutura de IF aninhado
nome = 'alex'
snome = 'silva'
idade = 36
if nome == 'alex' and snome == 'silva':
    if idade == 35:
        print('ok')
    else:
        print('ainda não')
else: 
    print('no way')
```

```python
# IF - com IN
nome = 'alexandre'
letra = 'j'
if letra in nome:
    print('ok')
else:
    print('não encontrado')
```

<font color='red'>- FOR

```python
# FOR - Estrutura LOOP FOR com range 
for i in range(0,10):
    print(i)
```

```python
# FOR - Estrutura LOOP FOR apresentando numeros pares, detalhe no modulo que é o resta da divisao
lt = [2,3,4,5,6,7,8]
for i in lt:
    if i %2 == 0:
        print(i)
```

```python
# FOR - Criando lista com o range e LOOP FOR
for i in range (0,10):
    lt.append(i)
print(lt)
lt.clear()
```

```python
# FOR - Criando lista e imprimindo numeros pares com modulo %
for i in range (0,10):
    lt.append(i)
for i in lt:
    if i %2 == 0:
        print(i)
lt.clear()
```

```python
# FOR - Criando lista e somando a variavel
for i in range (0,6):
    lt.append(i)
print(lt)
soma = 0
for i in lt:
    soma += i
print(soma)
lt.clear()
```

```python
# FOR - pulando uma etapa e continuando execução
for i in "python":
    if i == 'h':
        continue
    print(i)
```

```python
# FOR - Definindo numeros pares na lista
a = [1,2,3,4,5,6]
p = []
for i in a:
    if i %2 == 0:
        p.append(i)

print (p)
```

<font color='red'>- WHILE

```python
# WHILE - Estrutura
i = 0
while i < 10:
    print(i)
    i = i + 1
else:
    print('loop concluido')
```

```python
# WHILE - com if e break
i = 0
while i < 100:
    if i == 5:
        break
    else:
        pass
    print(i)
    i = i + 1
```

```python
# FUNCAO - Função lambda parametro de entrada e retorno - lambda x: xf
potencia = lambda num: num**2
potencia(5)
```

```python
# FUNCAO - Função lambada com string
palavra = lambda string: string[2]
palavra("python")
```

```python
# FUNCAO - Função Map com Lambda para somar matriz
a = [1,2,3,4]
b = [5,6,7,8]
list(map(lambda x,y:x+y,a,b))
```

```python
# FUNCAO - Função filter para numeros pares
def par(num):
    if num %2 == 0:
        return True
    else:
        return False

a = [1,2,3,4,5,6]
list(filter(par,a))
```

```python
#FOR E IF COMBINADOS
nome = 'alexandre'
letra = 'ale'
x = ''
for i in nome:
    if i in letra:
        pass
    else:
        x += '_ '
print (x)
```

```python
#FOR E IF COMBINADOS
nome = ['a','l','e']
letra = 'l'
x = ''
for i in nome:
    if i in letra:
        x += i+" "
    else:
        x += '_ '
print (x)
```
