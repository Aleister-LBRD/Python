```python
#Criando lista de listas
lista3 = [[1,2,3],[10,11,12]]
lista3 [0]
```

```python
#Criando variavel para extrair sublista
a = lista3 [0]
b = a[0]
b
```

```python
#Concatenar listas
lista_a = lista3 [0]
lista_b = lista3 [1]
lista_total = lista_a + lista_b
lista_total
```

```python
#Pesquisar itens em listas
print (10 in lista_total)
```

```python
#Tamanho da lista
len(lista_total)
```

```python
#Maior e Menor numero da lista
max(lista_total)
min(lista_total)
```

```python
#Adicionar e Remover itens a lista
lista_total.append(21)
lista_total.remove(21)
lista_total
```

```python
#ìndice da Lista para um elemento
lista_total.index (11)
```

```python
#inserir item na lista com indice
lista_total.insert(3, 23)
lista_total
```

```python
#Inverter lista
lista_total.reverse()
lista_total
```

```python
#Ordernar lista
lista_total.sort()
lista_total
```
